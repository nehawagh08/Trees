```python
#Q1 Implement Binary tree
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def addnode(self, data):
        if self.data:
            if data<self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.addnode(data)
            elif data>self.data:
                if self.right is None:
                    self.right=Node(data)
                else:
                    self.right.addnode(data)
        else:
            self.data = data
            
    def PrintTree(self):
        if self.left:
            self.left.PrintTree()
        print( self.data,end=" ")
        if self.right:
            self.right.PrintTree()
            
root = Node(12)
root.addnode(6)
root.addnode(14)
root.addnode(3)
root.PrintTree()
```

```python
#Q2 Find height of a given tree
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def addnode(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.addnode(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.addnode(data)
        else:
            self.data = data

    def height(self):
        if self is None:
            return 0
        left_height = 0
        right_height = 0
        if self.left:
            left_height = self.left.height()
        if self.right:
            right_height = self.right.height()
        return max(left_height, right_height) + 1

root = Node(12)
root.addnode(6)
root.addnode(14)
root.addnode(3)
print(root.height())
```

```python
#Q3 Perform Pre-order, Post-order, In-order traversal
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def addnode(self, data):
        if self.data:
            if data<self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.addnode(data)
            elif data>self.data:
                if self.right is None:
                    self.right=Node(data)
                else:
                    self.right.addnode(data)
        else:
            self.data = data
            
    def pre_order(self):
        print(self.data,end=" ")
        if self.left:
            self.left.pre_order()
        if self.right:
            self.right.pre_order()
            
    def post_order(self):
        if self.left:
            self.left.post_order()
        if self.right:
            self.right.post_order()
        print(self.data,end=" ")
    
    def in_order(self):
        if self.left:
            self.left.in_order()
        print(self.data,end=" ")
        if self.right:
            self.right.in_order()
            
root = Node(12)
root.addnode(6)
root.addnode(14)
root.addnode(3)
print("pre-order:")
root.pre_order()
print()
print("post-order:")
root.post_order()
print()
print("In-order:")
root.in_order()
```

```python
#Q5 Implement BFS (Breath First Search) and DFS (Depth First Search)
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class TreeTraversal:
    def __init__(self, root):
        self.root = root

    def bfs(self):
        if self.root is None:
            return

        queue = [self.root]

        while queue:
            node = queue.pop(0)
            print(node.data, end=' ')

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

    def dfs_preorder(self, node):
        if node is None:
            return

        print(node.data, end=' ')
        self.dfs_preorder(node.left)
        self.dfs_preorder(node.right)

    def dfs_inorder(self, node):
        if node is None:
            return

        self.dfs_inorder(node.left)
        print(node.data, end=' ')
        self.dfs_inorder(node.right)

    def dfs_postorder(self, node):
        if node is None:
            return

        self.dfs_postorder(node.left)
        self.dfs_postorder(node.right)
        print(node.data, end=' ')

# Example usage:
# Create a tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Create a TreeTraversal object
tree_traversal = TreeTraversal(root)

print("BFS traversal:")
tree_traversal.bfs()

print("\nDFS (preorder) traversal:")
tree_traversal.dfs_preorder(root)

print("\nDFS (inorder) traversal:")
tree_traversal.dfs_inorder(root)

print("\nDFS (postorder) traversal:")
tree_traversal.dfs_postorder(root)
```

```python
#Q6 Find sum of all left leaves in a given Binary Tree
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def addnode(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.addnode(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.addnode(data)
        else:
            self.data = data
    
    def left_sum(self, node):
        if node is None:
            return 0
        
        sum_left = 0
        
        if node.left is not None and node.left.left is None and node.left.right is None:
            sum_left += node.left.data
        
        sum_left += self.left_sum(node.left)
        sum_left += self.left_sum(node.right)

        return sum_left

root = Node(12)
root.addnode(6)
root.addnode(14)
root.addnode(3)
root.addnode(13)

print("Sum of left leaves in the given tree:")
print(root.left_sum(root))
```

```python
#Q7 Find sum of all nodes of the given perfect binary tree
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def perfect_binary_tree_sum(root):
    if root is None:
        return 0

    def calculate_height(node):
        height = 0
        while node:
            height += 1
            node = node.left
        return height

    def count_nodes_at_height(node, height):
        if height == 0:
            return 1
        if calculate_height(node.right) == height - 1:
            return 2 ** (height - 1) + count_nodes_at_height(node.right, height - 1)
        else:
            return 2 ** (height - 2) + count_nodes_at_height(node.left, height - 1)

    tree_height = calculate_height(root)
    total_nodes = count_nodes_at_height(root, tree_height)

    def calculate_sum(node, total_nodes):
        if node is None:
            return 0
        return node.value + calculate_sum(node.left, total_nodes // 2) + calculate_sum(node.right, total_nodes // 2)

    return calculate_sum(root, total_nodes)

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

sum_of_nodes = perfect_binary_tree_sum(root)
print("Sum of all nodes in the perfect binary tree:", sum_of_nodes)
```

```python
# Q8 Count subtress that sum up to a given value x in a binary tree
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def count_subtrees_with_sum(root, x):
    def dfs(node, x, count):
        nonlocal total_count

        if node is None:
            return 0

        left_sum = dfs(node.left, x, count)
        right_sum = dfs(node.right, x, count)

        subtree_sum = left_sum + right_sum + node.value

        if subtree_sum == x:
            total_count += 1

        return subtree_sum

    total_count = 0
    dfs(root, x, total_count)
    return total_count

root = TreeNode(5)
root.left = TreeNode(3)
root.right = TreeNode(7)
root.left.left = TreeNode(2)
root.left.right = TreeNode(4)
root.right.left = TreeNode(6)
root.right.right = TreeNode(8)

x = 12
count = count_subtrees_with_sum(root, x)
print(f"Number of subtrees with sum {x}: {count}")
```

```python
#Q9 Find maximum level sum in Binary Tree
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def max_level_sum(root):
    if root is None:
        return 0

    max_sum = float("-inf") 
    queue = [root]

    while queue:
        level_sum = 0
        level_size = len(queue)

        for i in range(level_size):
            node = queue.pop(0)
            level_sum += node.value

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        max_sum = max(max_sum, level_sum)

    return max_sum

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.right = TreeNode(6)

max_sum = max_level_sum(root)
print("Maximum level sum:", max_sum)
```

```python
#Q10 Print the nodes at odd levels of a tree
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def print_odd_level_nodes(root):
    def dfs(node, level):
        if node is None:
            return

        if level % 2 != 0:
            print(node.value, end=' ')

        dfs(node.left, level + 1)
        dfs(node.right, level + 1)

    dfs(root, 1)

# Example usage:
# Create a binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

print("Nodes at odd levels:")
print_odd_level_nodes(root)
```
